# PADRONIZAÇÃO DE COLUNAS PARA PORTUGUES

```python
# First, define a mapping for column name translation from English (dot-separated) to Portuguese
new_column_names = {
    'customerID': 'id_cliente',
    'Churn': 'evasao',
    'customer.gender': 'genero',
    'customer.SeniorCitizen': 'cidadao_senior',
    'customer.Partner': 'possui_parceiro',
    'customer.Dependents': 'possui_dependentes',
    'customer.tenure': 'meses_contrato',
    'phone.PhoneService': 'servico_telefone',
    'phone.MultipleLines': 'multiplas_linhas',
    'internet.InternetService': 'tipo_servico_internet',
    'internet.OnlineSecurity': 'seguranca_online',
    'internet.OnlineBackup': 'backup_online',
    'internet.DeviceProtection': 'protecao_dispositivo',
    'internet.TechSupport': 'suporte_tecnico',
    'internet.StreamingTV': 'streaming_tv',
    'internet.StreamingMovies': 'streaming_filmes',
    'account.Contract': 'tipo_contrato',
    'account.PaperlessBilling': 'fatura_online',
    'account.PaymentMethod': 'metodo_pagamento',
    'account.Charges.Monthly': 'cobranca_mensal',
    'account.Charges.Total': 'cobranca_total'
}

# Rename the columns in the DataFrame
df = df.rename(columns=new_column_names)

# Now, translate categorical values in the renamed columns

# Translate 'Yes'/'No' to 'Sim'/'Não' for relevant columns
yes_no_mapping = {'Yes': 'Sim', 'No': 'Não'}
for col in ['evasao', 'possui_parceiro', 'possui_dependentes', 'servico_telefone',
            'seguranca_online', 'backup_online', 'protecao_dispositivo',
            'suporte_tecnico', 'streaming_tv', 'streaming_filmes', 'fatura_online']:
    if col in df.columns:
        df[col] = df[col].replace(yes_no_mapping)

# Translate gender
df['genero'] = df['genero'].replace({'Female': 'Feminino', 'Male': 'Masculino'})

# Translate 'No phone service' and 'No internet service' and other specific terms
df['multiplas_linhas'] = df['multiplas_linhas'].replace({'No phone service': 'Sem serviço de telefone', 'Yes': 'Sim', 'No': 'Não'})
df['tipo_servico_internet'] = df['tipo_servico_internet'].replace({'No': 'Não', 'DSL': 'DSL', 'Fiber optic': 'Fibra óptica'})

# Handle 'No internet service' specifically for internet service columns
for col in ['seguranca_online', 'backup_online', 'protecao_dispositivo',
            'suporte_tecnico', 'streaming_tv', 'streaming_filmes']:
    if col in df.columns:
        df[col] = df[col].replace({'No internet service': 'Sem serviço de internet'})

# Translate contract type
df['tipo_contrato'] = df['tipo_contrato'].replace({
    'Month-to-month': 'Mensal',
    'One year': 'Anual',
    'Two year': 'Dois anos'
})

# Translate payment method
df['metodo_pagamento'] = df['metodo_pagamento'].replace({
    'Mailed check': 'Cheque enviado',
    'Electronic check': 'Cheque eletrônico',
    'Credit card (automatic)': 'Cartão de crédito (automático)',
    'Bank transfer (automatic)': 'Transferência bancária (automática)'
})

df.head()
```

3. CONTAGEM DE EVASÃO POR VARIAVEIS CATEGORICAS

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='genero', hue='evasao', palette='viridis')

plt.title('Contagem de Evasão por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem de Clientes')
plt.xticks(rotation=0)

churn_by_gender = df.groupby('genero')['evasao'].value_counts().unstack(fill_value=0)

churn_percentage_by_gender = churn_by_gender.apply(lambda x: x / x.sum() * 100, axis=1)

for patch in ax.patches:
    height = patch.get_height()
    if height == 0:
        continue

    x_center = patch.get_x() + patch.get_width() / 2
    # Ensure gender_index is within bounds of xticklabels
    gender_index = int(round(x_center))
    # Get the actual label from xticklabels
    if gender_index < len(ax.get_xticklabels()):
        gender_label = ax.get_xticklabels()[gender_index].get_text()
    else:
        # Handle cases where index might be out of bounds if plot is complex
        continue

    churn_value = None
    for i, container in enumerate(ax.containers):
        if patch in container.patches:
            # The churn_value is the index of the hue category in the container list
            # This assumes 'Não' is index 0 and 'Sim' is index 1 or vice-versa
            # We need to map the numerical churn_value to the string 'Não' or 'Sim'
            if i == 0:
                churn_str = 'Não'
            elif i == 1:
                churn_str = 'Sim'
            else:
                continue

            if churn_str in churn_percentage_by_gender.columns:
                percentage_val = churn_percentage_by_gender.loc[gender_label, churn_str]
                ax.annotate(f'{percentage_val:.1f}%', (x_center, height),
                            ha='center', va='bottom', fontsize=10, color='black')
            break

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))

# Define the palette colors explicitly
viridis_palette = sns.color_palette('viridis', n_colors=2) # Get 2 colors for 'Não' and 'Sim'

# Use catplot to create subplots for each gender
g = sns.catplot(data=df, x='possui_dependentes', hue='evasao', col='genero', kind='count', palette=viridis_palette, height=5, aspect=1.2)
g.set_axis_labels('Possui Dependentes', 'Contagem de Clientes')
g.set_titles('Evasão por Posse de Dependentes e Gênero: {col_name}')

# Calculate percentages for annotation
churn_by_dependents_gender = df.groupby(['genero', 'possui_dependentes'])['evasao'].value_counts().unstack(fill_value=0)

for ax in g.axes.flat:
    for p in ax.patches:
        height = p.get_height()
        if height == 0:
            continue

        # Get the gender label for the current subplot
        gender_label = ax.get_title().split(': ')[1]

        # Get the 'possui_dependentes' category label for the current bar
        # Use get_x() and get_width() to determine the bar's position and then map to xticklabels
        x_coord = p.get_x() + p.get_width() / 2
        dependents_label_str = ax.get_xticklabels()[int(round(x_coord))].get_text()

        # Determine churn category ('Não' or 'Sim') by comparing patch color to palette colors
        if p.get_facecolor() == viridis_palette[0]: # Assuming first color is 'Não'
            churn_str = 'Não'
        else: # Assuming second color is 'Sim'
            churn_str = 'Sim'

        # Retrieve percentage
        if dependents_label_str in churn_by_dependents_gender.loc[gender_label].index and churn_str in churn_by_dependents_gender.loc[gender_label].columns:
            total_for_category = churn_by_dependents_gender.loc[(gender_label, dependents_label_str)].sum()
            percentage = (churn_by_dependents_gender.loc[(gender_label, dependents_label_str), churn_str] / total_for_category) * 100
            ax.annotate(f'{percentage:.1f}%',
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
ax = sns.countplot(data=df, x='metodo_pagamento', hue='evasao', palette={'Não': 'red', 'Sim': 'blue'})

plt.title('Contagem de Evasão por Forma de Pagamento')
plt.xlabel('Forma de Pagamento')
plt.ylabel('Contagem de Clientes')
plt.xticks(rotation=45, ha='right')

payment_method_churn_counts = df.groupby('metodo_pagamento')['evasao'].value_counts().unstack(fill_value=0)
payment_method_percentage_by_churn = payment_method_churn_counts.apply(lambda x: x / x.sum() * 100, axis=1)

for patch in ax.patches:
    height = patch.get_height()
    if height == 0:
        continue

    x_center = patch.get_x() + patch.get_width() / 2
    payment_method_index = int(round(x_center))
    payment_method_label = ax.get_xticklabels()[payment_method_index].get_text()

    churn_str = None
    for i, container in enumerate(ax.containers):
        if patch in container.patches:
            # Assuming 'Não' is the first category and 'Sim' is the second
            if i == 0:
                churn_str = 'Não'
            elif i == 1:
                churn_str = 'Sim'
            break

    if churn_str is not None and payment_method_label in payment_method_percentage_by_churn.index:
        if churn_str in payment_method_percentage_by_churn.columns:
            percentage_val = payment_method_percentage_by_churn.loc[payment_method_label, churn_str]
            ax.annotate(f'{percentage_val:.1f}%', (x_center, height),
                        ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

A análise foi dividida em três dimensões:

- Relação com variáveis numéricas  

---

## 3.1 Distribuição Geral do Churn

A base apresenta taxa relevante de cancelamento, com leve desbalanceamento entre classes — comportamento comum em datasets do setor de telecomunicações.

Esse cenário reforça a necessidade de estratégias direcionadas à retenção.

---

## 3.2 Análise por Variáveis Categóricas

### Tipo de Contrato

Clientes com contrato **Month-to-month** apresentam a maior taxa de churn.  

Contratos **One-year** e **Two-year** demonstram retenção significativamente superior.

**Interpretação:** a ausência de fidelização reduz a barreira de saída e aumenta a volatilidade da carteira.

---

### Tipo de Serviço de Internet

Clientes com **Fiber Optic** apresentam churn superior aos clientes com **DSL**.

Possível explicação: planos de fibra possuem mensalidade mais elevada, aumentando a sensibilidade ao preço.

---

### Serviços Adicionais

A ausência de:

- OnlineSecurity  
- TechSupport  
- DeviceProtection  

está fortemente associada ao cancelamento.

Clientes com maior diversificação de serviços demonstram maior retenção, possivelmente devido ao aumento do custo de troca e maior percepção de valor agregado.

---

### Método de Pagamento

O método **Electronic Check** concentra parcela significativa dos cancelamentos.

Esse padrão pode indicar menor automatização do pagamento e maior volatilidade no comportamento do cliente.

---

## 3.3 Análise por Variáveis Numéricas

### Tempo de Permanência (Tenure)

Clientes que cancelam possuem tempo médio de permanência significativamente menor.

**Conclusão:** os primeiros meses representam o período de maior risco de churn.

---

### Cobrança Mensal (Monthly Charges)

Clientes churn apresentam mensalidade média superior à dos clientes ativos.

Isso indica sensibilidade ao preço e possível desalinhamento entre custo e percepção de benefício.

---

### Total Gasto (Total Charges)

Clientes churn apresentam menor valor acumulado, reforçando que grande parte dos cancelamentos ocorre de forma precoce.

---

### Contas Diárias

Clientes churn apresentam custo diário ligeiramente maior, corroborando a hipótese de desconforto financeiro.

---

## 4. Principais Padrões Identificados

A evasão na TelecomX está fortemente associada a:

- Contratos de curta duração  
- Baixo tempo de permanência  
- Mensalidades elevadas  
- Ausência de serviços adicionais  
- Método de pagamento Electronic Check  
- Serviço de internet Fibra Óptica  

Os padrões indicam concentração do risco em clientes recém-adquiridos e com menor vínculo contratual.

O churn identificado não ocorre de forma aleatória, mas está estruturalmente relacionado a variáveis contratuais e comportamentais.

A empresa apresenta maior vulnerabilidade:

- Nos primeiros meses de relacionamento  
- Em clientes com maior custo mensal  
- Em contratos sem fidelização
